```python
import json
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

## Radial encoding

```python
file_name = "results_52821"

with open(f"results/{file_name}.json", "r") as fp:
    results = json.load(fp)

print("Ran on:", results["time"])
for key in ["number_of_features", "regularization", "train_acc", "test_acc"]:
    print(f"{key}:", results[key])
```

```python
y_true = np.array(results["labels_true"])
y_pred = np.array(results["labels_pred"])
(y_true == y_pred).mean()
```

```python
results["label_names"].values()
```

```python
conf_mat = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(conf_mat, display_labels=results["label_names"].values())
disp.plot()
plt.show()
```

## Convolutional encoding
